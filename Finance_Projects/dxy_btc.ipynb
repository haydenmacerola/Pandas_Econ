# Assessing the relationship between $DXY and $BTC 

```python
import yfinance as yf
import matplotlib.pyplot as plt 
import pandas as pd 
import datetime as dt
```

```python
dxy = yf.Ticker("DX-Y.NYB")
dxy.fast_info


```

```python
dxy_5yr = dxy.history(period="5y")
dxy_5yr = dxy_5yr['Open']
dxy_5yr = dxy_5yr.to_frame()
dxy_5yr.head()
```

```python
btc = yf.Ticker("BTC-USD")
btc.fast_info
```

```python
btc_5yr = btc.history(period="5y")
btc_5yr = btc_5yr['Open']
btc_5yr = btc_5yr.to_frame()
btc_5yr.head()
```

```python
btc_5yr.plot()
```

```python
dxy_5yr.plot()
```

```python
fig, ax1 = plt.subplots()
ax1.plot(btc_5yr, label= '$BTC', color = 'orange')
ax1.set_ylabel('BTC $USD')
ax1.set_xlabel('Date')
ax2 = ax1.twinx()
ax2.plot(dxy_5yr, color = 'lime', label='$DXY')
ax2.set_ylabel('DXY $USD ')
fig.legend(loc='upper left', bbox_to_anchor=(0.2,0.8))
plt.title('DXY v BTC open price 2018 - 2023')
```
